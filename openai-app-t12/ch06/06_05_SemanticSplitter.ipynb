```python
import os
from dotenv import load_dotenv

load_dotenv("../../_apikeys.env")
api_key = os.getenv("DoogieOpenaiKey")
os.environ['OPENAI_API_KEY'] = api_key
```

```python
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/chatgpt-kr/openai-api-tutorial/main/ch06/test.txt", filename="06_05_test.plaintext" )
with open("06_05_test.plaintext", encoding="utf-8") as f:
    textfile = f.read()
print("텍스트길이:", len(textfile) )
```

```python
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_experimental.text_splitter import SemanticChunker

embeddings = OpenAIEmbeddings()
textSplitter = SemanticChunker( embeddings )
#textSplitter = SemanticChunker( OpenAIEmbeddings() )

textlist = [ textfile ]
splitedlist = textSplitter.create_documents( textlist )
#splitedlist = textSplitter.create_documents( [ textfile] )
```

```python
print("current OpenAIEmbeddings().model=", embeddings.model)
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
textSplitter = SemanticChunker( embeddings )
print("current OpenAIEmbeddings().model=", embeddings.model)
```

```python
print( type(textfile), type(textlist), type(splitedlist) )
print( len(textfile),  len(textlist),  len(splitedlist) )
print( type(splitedlist[0]), type(splitedlist[0].page_content), len(splitedlist[0].page_content) )
print( splitedlist[0].page_content[:30], "\n", splitedlist[1].page_content[:30] )
print( splitedlist[0].type, splitedlist[1].type )
```

```python
for i in range(len(splitedlist)) :
    print( "splitedlist[{}]".format( i ), len(splitedlist[i].page_content),
           splitedlist[i].page_content[:30] )
```

```python
textsplitter = SemanticChunker(
    OpenAIEmbeddings(),
    breakpoint_threshold_type="percentile",
    breakpoint_threshold_amount=95,
)

splitedlist = textsplitter.create_documents( [textfile] )
print( type(textfile), type(textlist), type(splitedlist), )
print( len(textfile), len(textlist), len(splitedlist), )
```

```python
for i in range(len(splitedlist)) :
    print( "splitedlist[{}]".format( i ), len(splitedlist[i].page_content),
           splitedlist[i].page_content[:30] )
```
