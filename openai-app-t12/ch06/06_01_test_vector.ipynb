```python
import numpy as np
#from langchain_openai import OpenAIEmbeddings
from numpy import dot
from numpy.linalg import norm

def cosine_similarity(a, b):
    print( dot(a,b) )
    print( norm(a) )
    print( norm(b) )
    print( norm(a) * norm(b))
    return dot(a, b) / (norm(a) * norm(b))

vec1 = np.array( [0,1,1,1] )
vec2 = np.array( [1,0,1,1] )
vec3 = np.array( [2,0,2,2] )

print("vec 1,2 sim = ", cosine_similarity(vec1, vec2))
print("vec 1,3 sim = ", cosine_similarity(vec1, vec3))
print("vec 2,3 sim = ", cosine_similarity(vec2, vec3))
```

```python
vec1 = np.array( [0,2] )
vec2 = np.array( [-1,-1] )
vec3 = np.array( [2,2] )
vec4 = np.array( [3,3] )
print("vec 1,2 sim = ", cosine_similarity(vec1, vec2))
print("vec 1,3 sim = ", cosine_similarity(vec1, vec3))
print("vec 1,4 sim = ", cosine_similarity(vec1, vec4))
print("vec 3,4 sim = ", cosine_similarity(vec3, vec4))
```
