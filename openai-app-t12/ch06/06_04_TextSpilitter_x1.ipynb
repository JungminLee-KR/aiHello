```python
import urllib.request
from sys import getsizeof

urllib.request.urlretrieve("https://raw.githubusercontent.com/lovit/soynlp/master/tutorials/2016-10-20.txt", filename="06_04_test.plaintext" )
```

```python
with open("06_04_test.plaintext", encoding="utf-8") as f:
    textfile = f.read()

print( "A", type(textfile), len(textfile) )
print( "AA", textfile[0:30] )
print( "B", type([textfile]), len([textfile]), len( [textfile][0]) )
print( "BB", [textfile][0][0:30])
```

```python
# 문서를 만들어 벡터 DB에 넣거나, RAG에서 활용하고 싶다면,
# split_text()가 아니라 create_documents()를 사용하여야 한다.
# create_documents()는 argument로 여러개의 text list를 입력으로 받는 것에 주의 한다.
# 하나의 text만 취급할때는 그 text를 리스트로 감싸서
# 원소가 하나인 리스트 타입으로 만들어 넘겨 주면 된다.
#
from langchain.text_splitter import RecursiveCharacterTextSplitter

textSplitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=0,
    is_separator_regex=False,
    length_function=len,)

textlist = [ textfile ]
splitedlist = textSplitter.create_documents( textlist )

#for doc in texts1:
#    print(doc.page_content, doc.metadata)
```

```python
print( "AA", type(textfile), len(textfile) )
print( "AB", type(textlist), len(textlist) )
print( "AC", type(textlist[0]), len(textlist[0]) )
print( "BA", type(splitedlist), len(splitedlist) )
print( "BB", type(splitedlist[0]), sys.getsizeof(splitedlist[0]), splitedlist[0].type )
print( "CA", type(splitedlist[0].page_content), len(splitedlist[0].page_content) )
print( "CB", splitedlist[0].page_content[:30] )
print( "CC", splitedlist[1].page_content[:30] )
```

```python
print( "A", textfile[:40] )
print( "B", textlist[0][:40] )
print( "C", splitedlist[0].page_content[:40] )
print( "D", splitedlist[1].page_content[:40] )
```

```python
for i in range( 30 ):
    print ("[",i,"]=",textlist[ 0 ][i], type(textlist[ 0 ][i]), )
for i in range( 30 ):
    print ("[",i,"]=",textfile[i], type(textfile[i]), )
```

```python
for i in range(10):
    print( "[", i, "]=", len(splitedlist[i].page_content), "::", splitedlist[i].page_content[:30] )
```
