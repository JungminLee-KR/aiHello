```python
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/lovit/soynlp/master/tutorials/2016-10-20.txt", filename="06_04_test.plaintext" )
```

```python
with open("06_04_test.plaintext", encoding="utf-8") as f:
    textfile = f.read()
print("텍스트길이:", len(textfile) )

print( "A", type(textfile), len(textfile), textfile[0:30] )
print( "C", type([textfile]), len([textfile]), len( [textfile][0]), [textfile][0][0:30])
```

```python
# 문서를 만들어 벡터 DB에 넣거나, RAG에서 활용하고 싶다면,
# split_text()가 아니라 create_documents()를 사용하여야 한다.
# create_documents()는 여러개의 text list를 입력으로 받는 것에 주의 한다.
# 하나의 text만 취급할때는 그 text를 리스트로 감싸서
# 원소가 하나인 리스트 타입으로 만들어 넘겨 주면 된다.
#
from langchain.text_splitter import RecursiveCharacterTextSplitter

textSplitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=0,
    is_separator_regex=False,
    length_function=len,)

textlist = [ textfile ]
splitedlist = textSplitter.create_documents( textlist )

#for doc in texts1:
#    print(doc.page_content, doc.metadata)
```

```python
print( type(textfile), type(textlist), type(splitedlist), type(splitedlist[0]) )
print( len(textfile),  len(textlist),  len(splitedlist) )
print( type(splitedlist[0].page_content), len(splitedlist[0].page_content) )
print( splitedlist[0].page_content[:30], "\n", splitedlist[1].page_content[:30] )
print( splitedlist[0].type, splitedlist[1].type )
```

```python
print( textfile[:40] )
print( textlist[0][:40] )
```

```python
for i in range( 30 ):
    print ("[",i,"]=",textlist[ 0 ][i], type(textlist[ 0 ][i]), )
for i in range( 30 ):
    print ("[",i,"]=",textfile[i], type(textfile[i]), )
```

```python
for i in range(10):
    print( "[", i, "]=", len(splitedlist[i].page_content), "::", splitedlist[i].page_content[:30] )
```
