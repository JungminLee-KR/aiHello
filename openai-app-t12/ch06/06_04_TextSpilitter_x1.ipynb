```python
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/lovit/soynlp/master/tutorials/2016-10-20.txt", filename="06_04_test.plaintext" )
```

```python
with open("06_04_test.plaintext", encoding="utf-8") as f:
    textfile = f.read()
print("텍스트길이:", len(textfile) )
```

```python
# 문서를 만들어 벡터 DB에 넣거나, RAG에서 활용하고 싶다면,
# split_text()가 아니라 create_documents()를 사용하여야 한다.
# create_documents()는 여러개의 text list를 입력으로 받는 것에 주의 한다.
# 하나의 text만 취급할때는 그 text를 리스트로 감싸서 리스트로 만들어 넘겨주면 된다.
#
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitedText = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=0,
    is_separator_regex=False,
    length_function=len,)

textlist = [ textfile ]
splittexts = splitedText.create_documents( textlist )

#for doc in texts1:
#    print(doc.page_content, doc.metadata)
```

```python
print( type(textfile), type(textlist), type(textlist[0]),
       type(splittexts), type(splittexts[0].page_content) )
print( len(textfile), len(textlist), len(textlist[0]),
        len(splittexts), len(splittexts[0].page_content) )
```
