```python
#
# Hello
# allow to in-OpenAi_Azure at office(12.4.80.170)
####샘플코드
from openai import AzureOpenAI
from dotenv import load_dotenv
import os

load_dotenv(override=True)
load_dotenv("../../_apikeys.env")

#azure endpoint 및 api키, 버전, 모델정보 가져오기
azure_endpoint=os.getenv("AZURE_OPENAI_4.1_ENDPOINT","")
api_key = os.getenv("AZURE_OPENAI_4.1_API_KEY")
api_version = os.getenv("OPENAI_API_4.1_VERSION")
deployment_name    = os.getenv('DEPLOYMENT_4.1_NAME')

print("azure_endpoint:", azure_endpoint)
print("api_key:", api_key )
print("api_version:", api_version )
print("deployment_name:", deployment_name )

#클라이언트 설정
openai_client = AzureOpenAI(
    azure_endpoint = azure_endpoint,
    api_key        = api_key,
    api_version    = api_version
)

# 예시 테스트: 모델 리스트 가져오기
try:
    response = openai_client.models.list()
    print("클라이언트 생성 및 API 통신 성공!, Model List:")
    #print("모델 리스트 예시:", [model.id for model in response.data])  # 일부만 출력
    for model in sorted(response.data, key=lambda model:model.id):
        print(model.id)
except Exception as e:
    print("클라이언트 생성 또는 API 호출 실패:", e)
```
