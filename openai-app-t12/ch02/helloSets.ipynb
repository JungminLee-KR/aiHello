```python
# tuple, can not change
atuple = (1,2,3)
print( f"tuple ()...: {type(atuple)}" )
#atuple[0]="11s"
for aa in atuple:
    print(aa, type(aa), end=' ')
print(len(atuple))
```

```python
# list (as array)
alist = [1,2,3]
print( alist )
print( f"list []...: {type(alist)}" )
alist[0]="11s"
for aa in alist:
    print(aa, type(aa), end=' ')
print(len(alist))
```

```python
# dictionary
adict ={}
print( adict.items() , len(adict))
adict = {'AAA':1, 'BBB':2, 'CCC':3}
print( adict.items() , len(adict))
print( f"dictionary []...: {type(adict)}, {len(adict)}" )
for aa in adict:
    print(aa,":", adict[aa], ">", type(adict[aa]))
print(len(adict))

adict['BBB']="new11s"
for aa in adict:
    print(aa,":", adict[aa], ">", type(adict[aa]))
print(len(adict))
```
